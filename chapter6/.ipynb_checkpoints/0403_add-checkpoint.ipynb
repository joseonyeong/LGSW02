```python
# os module ; operating System
# 파일과 관련된 기능을 제공
import os
# glob module
# unix에서 사용하는 규칙에 따라 패턴과 일치하는 모든 경로 이름을 찾음
# 파일명을 리스트 형식으로 반환해줌.
import glob

# .txt 확장자인 파일을 찾아서 file에 넣어
for file in glob.glob(os.path.join('*.txt')):
    # 읽기 모드로 열어라
    with open(file, 'r', newline = '', encoding = 'utf-8') as filereader:
        # 행마다 받고
        for row in filereader:
            # 옆 공백 제거하고 포매팅
            print("{}".format(row.strip()))
```

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
pivot_df = pd.pivot_table(
    tips,
    values = 'tip', # 집계할 값
    index = 'sex', # 행
    columns = 'smoker', # 열
    aggfunc = 'mean', # 집계함수
    observed = True
)
pivot_df
```

```python
result1 = tips.groupby(['sex', 'smoker'], observed = True)['tip'].mean()
```

```python
result2 = tips.groupby(['sex', 'smoker'], observed = True)['tip'].mean().unstack()
```
